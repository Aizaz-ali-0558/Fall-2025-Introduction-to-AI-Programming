# **Lab Report Task:**
You are developing a **simple energy consumption logging system** for a residential building.
Daily electricity usage is stored in a **CSV file**. so it can be easily read, updated, and organized.



_________________

## **Task 1:**
### **Create and Write Energy Data to a CSV File**
1. Create a file named energy_data.csv.<br>
Write the header and daily energy consumption records into the file using the CSV format:
-	Date
-	Energy consumed in kWh
Each record must be written on a new line following the header.


```python
import csv

# Data to write
energy_data = [
    ["2025-01-01", 12.5],
    ["2025-01-02", 14.2],
    ["2025-01-03", 11.8],
    ["2025-01-04", 13.6],
    ["2025-01-05", 15.1]
]
```

```python
# Create and write to CSV file
with open("energy_data.csv", mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Date", "Energy consumed in kWh"]) # Write header
    writer.writerows(energy_data) # Write data rows

print("energy_data.csv file created successfully!")
```

________________

## **Task 2:**
###  **Read and Display CSV File Data**
Read the contents of energy_data.csv and display all energy consumption records in a clear and readable format.
- The data must be read from the file, not hardcoded.


```python
import csv

with open("energy_data.csv", "r") as file:
    reader = csv.reader(file)
    header = next(reader) # Skip header
    for row in reader: # Display records
        print(f"Date: {row[0]}  Energy Consumed: {row[1]} kWh")
```

## **Task 3:**
### **Append New Energy Records**
Add new daily energy consumption records to energy_data.csv without removing existing data.
Ensure:
-	The CSV format is preserved
-	New data is appended correctly


```python
import csv

with open("energy_data.csv", "a", newline="") as file:
    writer = csv.writer(file)

    # Append new records
    writer.writerow(["2025-01-05", 13])
    writer.writerow(["2025-01-06", 14])
```
