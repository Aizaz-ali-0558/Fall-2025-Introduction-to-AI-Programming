You are working in a university electronics lab. <br>
You are asked to monitor a small solar home system installed on the lab roof. The system has:<br>
a)	A solar panel<br>
b)	A battery<br>
c)	A load controller<br>
d)	A temperature sensor<br>
Each minute, the system sends a small dictionary of readings<br>
I.	Create a dictionary named reading1 containing the 5 fields shown in the scenario.<br>
II.	From reading1, print:<br>
i.	The solar power<br>
ii.	The battery voltage<br>
iii.	The load status<br>
III.	Print how many fields the dictionary contains after your modifications.<br>
IV.	Remove the "panel_current" field.<br>
V.	Get Keys, Values, and Items<br>
**Print:**<br>
i.	All keys<br>
ii.	All values<br>
iii.	All key-value pairs<br>
iv.	Use .keys(), .values(), and .items().<br>



```python
reading1 = {
    "solar_power": 120.5,      
    "battery_voltage": 12.8,   
    "panel_current": 5.4,        
    "load_status": "ON",        
    "temperature": 32.5          
}
```

```python
print("Solar Power:", reading1["solar_power"])
print("Battery Voltage:", reading1["battery_voltage"])
print("Load Status:", reading1["load_status"])
```

```python
print("Number of fields:", len(reading1))
```

```python
reading1.pop("panel_current")
```

```python
print("Number of fields after removal:", len(reading1))
```

```python
print("All Keys:", reading1.keys())
print("All Values:", reading1.values())
print("All Key-Value Pairs:", reading1.items())
```
