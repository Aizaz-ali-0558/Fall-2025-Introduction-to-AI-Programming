You are helping a small learning center clean and organize small classroom datasets. They give you short dictionaries and want quick transformations using comprehensions only, because they want the code short and readable. Create dictionary from a list.

### **Given:**
names = ["Ali", "Sara", "Bilal"]
Create a dictionary where each name maps to its length.
{key: value for item in iterable}
Filter a dictionary using comprehension

```python
names = ["Ali", "Sara", "Bilal"]
len_name={x:len(x) for x in names}
len_name
```

## **Given:**
scores = {"Ali": 55, "Sara": 82, "Bilal": 40, "John": 90}<br>
Create a new dictionary containing only the students who scored 60 or above.<br>
{k: v for k, v in dict.items() if condition}<br>
Swap keys and values<br>

```python
scores = {"Ali": 55, "Sara": 82, "Bilal": 40, "John": 90}
student_above_60={k:v for k,v in scores.items() if v>60}
student_above_60
```

## **Given:**
pairs = {"a": 1, "b": 2, "c": 3}<br>
Use a dictionary comprehension to swap keys and values.<br>
Transform dictionary values<br>

```python
pairs = {"a": 1, "b": 2, "c": 3}
sweep_k_v={k:v for v,k in pairs.items() }
sweep_k_v
```

### **Given:**
prices = {"pen": 10, "book": 50, "bag": 100}<br>
Use a dictionary comprehension to apply a 10% discount to each price.

```python
prices = {"pen": 10, "book": 50, "bag": 100}
discount_prices={k:((v/100)*10) for k,v in prices.items()}
discount_prices
```
