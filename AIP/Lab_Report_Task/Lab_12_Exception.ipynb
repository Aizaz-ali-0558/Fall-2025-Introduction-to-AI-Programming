# **Lab Report Task:**
You are developing a Student Result Processing System for a small institute.<br>
The program takes student marks as input and calculates results. Since users may enter incorrect data, the program must handle errors properly using Python exception handling.


____________

## **Task 1:**
### **Handle Invalid User Input**
**Problem**<br>
The program asks the user to enter numeric values for total marks and obtained marks. Users may enter invalid input such as letters or symbols.<br>
The program must not crash and should display a clear error message.<br>
**Use the Following Concepts**
-	try
-	except
-	ValueError


```python
try:
    total_marks = float(input("Enter total marks: "))
    obtained_marks = float(input("Enter obtained marks: "))

    percentage = (obtained_marks / total_marks) * 100
    print("Percentage:", percentage)

except ValueError:
    print("Error: Please enter numeric values only.")
```

______________

## **Task 2:**
### **Handle Division by Zero**
#### **Problem**
If total marks are entered as zero, calculating the percentage will result in a runtime error.<br>
Modify the program to safely handle this case and notify the user.<br>
**Use the Following Concepts**
-	ZeroDivisionError
-	Multiple except blocks


```python
try:
    obtained_marks = float(input("Enter obtained marks: "))
    total_marks = float(input("Enter total marks: "))

    percentage = (obtained_marks / total_marks) * 100
    print("Percentage:", percentage)

except RuntimeError as RT_Error:
    print(RT_Error)

except ZeroDivisionError as ZD_Error:
    print(ZD_Error)
```

________________

## **Task 3:**
### **Validate Logical Conditions Using raise**
**Problem**
Obtained marks must not be negative and must not exceed total marks.<br>
If the condition is violated, the program should manually raise an exception with a meaningful message.<br>
**Use the Following Concepts**
-	raise
-	Exception


```python
obtained = float(input("Enter obtained marks: "))
total = float(input("Enter total marks: "))

if obtained < 0 or obtained > total:
    raise Exception("Obtained marks must be between 0 and total marks.")

print("Valid marks entered.")
```

## **Task 4:**
### **Execute Code Only When No Error Occurs**
**Problem**
Display the calculated percentage and pass or fail status only if no exception occurs during input and validation.

***Use the Following Concepts***
-	try
-	except
-	else


```python
try:
    total = float(input("Enter total marks: "))
    obtained = float(input("Enter obtained marks: "))

    if obtained < 0 or obtained > total:
        raise Exception("Invalid marks")

    percentage = (obtained / total) * 100

except Exception as e:
    print("Error:", e)

else:
    print("Percentage:", percentage)

    if percentage >= 50:
        print("Status: Pass")
    else:
        print("Status: Fail")
```

## **Task 5:**
### **Guarantee Program Completion Using finally**
**Problem**<br>
No matter what happens during execution, the program must display:<br>
Result processing completed.<br>
This ensures proper program termination behavior.<br>
Use the Following Concepts<br>
-	finally


```python
try:
    total = float(input("Enter total marks: "))
    obtained = float(input("Enter obtained marks: "))

    if obtained < 0 or obtained > total:
        raise Exception("Invalid marks")

    percentage = (obtained / total) * 100
    print("Percentage:", percentage)

except Exception as e:
    print("Error:", e)

finally:
    print("Result processing completed.")
```
