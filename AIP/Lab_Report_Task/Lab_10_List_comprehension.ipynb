### **Task 1:**  Basic list comprehension
Create a new list of numbers from 0 to 9 using a list comprehension, not a loop.<br>
basic syntax [expression for item in iterable]

```python
numbers=[x for x in range(10)]
numbers
```

______________

### **Task 2:** Filtering items with if
Given:<br>
nums = [1, 2, 3, 4, 5, 6]<br>
Use a list comprehension with an if clause to keep only the even numbers.<br>
 [expr for item in iterable if condition]<br>

```python
nums=[1,2,3,4,5,6]
even_nums=[x for x in nums if x%2==0]
even_nums
```

__________________

### **Task 3:**  Conditional expression inside comprehension
Given:<br>
nums = [1,2,3,4]<br>
Create a new list where even numbers become "even" and odd numbers become "odd".<br>
expr_if_true if condition else expr_if_false inside comprehension<br>

```python
nums=[1,2,3,4]
even_and_odd = ["even" if x % 2 == 0 else "odd" for x in nums]
even_and_odd
```

___________

### **Task 4:** Simple nested list comprehension
Given a list of lists:<br>
data = [[1,2], [3,4]]<br>
Flatten it into a single list using a nested list comprehension.<br>


```python
data=[[1,2],[3,4]]
one_list_data=[x for s in data for x in s]
one_list_data
```
