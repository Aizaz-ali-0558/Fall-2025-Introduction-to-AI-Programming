```python
# Function to calculate total power for each key
def total_power(data_dict):
    result = {}
    for key, values in data_dict.items():
        result[key] = sum(values)
    return result
```

This function is used to aggregate energy consumption data by summing multiple readings into a single total for each time category.

```python
# Function to calculate weekday or weekend consumption
def days_consumption(day_wise_power, *days):
    total = 0
    for day, power in day_wise_power.items():
        if day in days:
            total += power
    return total
```

This function calculates total energy consumption for selected days by summing their daily values, allowing comparison between weekday and weekend usage.

```python
import csv

month_dict = {}
day_dict = {}
hour_dict = {}

with open("Building_2.csv", "r") as file:
    reader = csv.reader(file)

    for i, row in enumerate(reader):
        if i == 0:
            continue  # skip header

        month = int(float(row[0]))
        hour = int(float(row[1]))
        day = int(float(row[2]))
        power = float(row[7])

        ''' ‚ÄúIf available‚Äù means checking whether the key already exists in the dictionary.
          If it exists, the value is appended; 
          otherwise, a new key with an empty list is created.'''
        
        # Month wise
        if month not in month_dict:
            month_dict[month] = []
        month_dict[month].append(power)

        # Day wise
        if day not in day_dict:
            day_dict[day] = []
        day_dict[day].append(power)

        # Hour wise
        if hour not in hour_dict:
            hour_dict[hour] = []
        hour_dict[hour].append(power)
```

```python
'''This code loops through all monthly power values,
 calculates their sums, and adds them together to
   obtain the total annual energy consumption.'''

annual_consumption = 0
for values in month_dict.values():
    annual_consumption += sum(values)

print("Total Annual Consumption:", annual_consumption)
```

```python
'''This code uses a function to calculate monthly energy consumption
 and prints it in ascending order of months for better readability.'''

print("Monthly Consumption:")
monthly_power = total_power(month_dict)

for month, power in sorted(monthly_power.items()):
    print(f"Month {month}: {power}")
```

```python
print("Daily Consumption:")
day_wise_power = total_power(day_dict)

for day, power in sorted(day_wise_power.items()):
    print(f"Day {day}: {power}")
```

```python
print("Hourly Consumption:")
hourly_power = total_power(hour_dict)

for hour, power in sorted(hourly_power.items()):
    print(f"Hour {hour}: {power}")
```

```python
'''Seasonal consumption is calculated by summing the energy values
 of the months that belong to each season.'''
# inner sum: Adds all power values of one month
# outer sum: Adds totals of all selected months
winter = sum(sum(v) for m, v in month_dict.items() if m in [12, 1, 2])
spring = sum(sum(v) for m, v in month_dict.items() if m in [3, 4, 5])
summer = sum(sum(v) for m, v in month_dict.items() if m in [6, 7, 8])
autumn = sum(sum(v) for m, v in month_dict.items() if m in [9, 10, 11])

print("\nSeasonal Consumption:")
print("Winter:", winter)
print("Spring:", spring)
print("Summer:", summer)
print("Autumn:", autumn)
```

```python
'''This code compares weekday and weekend energy consumption
 by calculating their total and average daily usage.'''

weekday_consumption = days_consumption(day_wise_power, 1, 2, 3, 4, 5)
weekend_consumption = days_consumption(day_wise_power, 6, 7)

print("\nWeekday Consumption:", weekday_consumption)
print("Weekend Consumption:", weekend_consumption)

# Comparison
print("\nAverage Consumption Comparison:")
if weekend_consumption / 2 > weekday_consumption / 5:
    print("Weekend average consumption is higher")
else:
    print("Weekday average consumption is higher")
```

## **Visualization**

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (9, 5)
plt.rcParams["font.size"] = 11
```

```python
months = sorted(monthly_power.keys())
month_values = [monthly_power[m] for m in months]

plt.figure()
plt.bar(months, month_values)
plt.xlabel("Month")
plt.ylabel("Energy Consumption (kWh)")
plt.title("Monthly Energy Consumption of Building 2")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
hours = sorted(hourly_power.keys())
hour_values = [hourly_power[h] for h in hours]

plt.figure()
plt.plot(hours, hour_values, linewidth=2, marker='o')
plt.xlabel("Hour of Day")
plt.ylabel("Energy Consumption (kWh)")
plt.title("Hourly Energy Consumption Pattern")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
seasons = ["Winter", "Spring", "Summer", "Autumn"]
season_values = [winter, spring, summer, autumn]

plt.figure()
plt.bar(seasons, season_values)
plt.xlabel("Season")
plt.ylabel("Energy Consumption (kWh)")
plt.title("Seasonal Energy Consumption")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
labels = ["Weekdays (Mon‚ÄìFri)", "Weekend (Sat‚ÄìSun)"]
values = [weekday_consumption, weekend_consumption]

plt.figure()
plt.bar(labels, values)
plt.ylabel("Total Energy Consumption (kWh)")
plt.title("Weekday vs Weekend Energy Consumption")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

The graphs clearly show that energy consumption varies significantly across time.
Monthly and seasonal plots indicate higher usage during summer months due to cooling demand.
Hourly analysis reveals peak consumption during evening hours (17‚Äì21).
Although weekday total consumption is higher, weekend average daily consumption is greater.
