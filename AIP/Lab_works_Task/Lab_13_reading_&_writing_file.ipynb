## **Load Demand Data Recording**
An electrical feeder‚Äôs hourly load demand (kW) needs to be recorded for monitoring.<br>
________________
## **Tasks:**
-	Create a CSV file to store hourly load data
-	Write load values with time stamps
-	Organize each record on a separate line


```python
import csv

with open("load_demand.csv", "w", newline="") as file:
    writer = csv.writer(file)

    # Header
    writer.writerow(["Time", "Load Demand (kW)"])

    # Hourly records
    writer.writerow(["01:00", 120])
    writer.writerow(["02:00", 115])
    writer.writerow(["03:00", 110])
    writer.writerow(["04:00", 130])
```

## **Solar Energy Generation Logging**
A solar PV system records daily energy generation (kWh).<br>

## **Substation Fault Event Logging**
A substation records fault events during operation.<br>
## **Tasks:**
-	Create a fault log file
-	Write initial fault event records
-	Append future fault events as they occur


```python
import csv

# Create file and write initial fault event records
with open("fault_log.csv", "w", newline="") as file:
    writer = csv.writer(file)
    
    # Header
    writer.writerow(["Date", "Time", "Fault Type", "Equipment", "Status"])
    
    # Initial fault events
    writer.writerow(["2025-01-05", "10:15", "Overcurrent", "Transformer T1", "Cleared"])
    writer.writerow(["2025-01-06", "14:30", "Earth Fault", "Busbar B2", "Cleared"])
```

```python
# Append future fault events
with open("fault_log.csv", "a", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["2025-01-07", "09:45", "Overvoltage", "Line L3", "Cleared"])
    writer.writerow(["2025-01-08", "16:20", "Short Circuit", "Transformer T2", "Investigating"])
```

## **Task:**
### **Organize and Process CSV Data**
Read the CSV file line by line and:<br>
-	Count the total number of recorded days
-	Display each record with a serial number<br>

The header must not be counted as a data record.


```python
import csv

with open("energy_data.csv", "r") as file:
    reader = csv.reader(file)
    
    # Skip header
    header = next(reader)
    
    # Initialize counter
    total_days = 0
    
    # Display records with serial number
    for idx, row in enumerate(reader, start=1):
        print(f"{idx}. Date: {row[0]} | Energy Consumed: {row[1]} kWh")
        total_days += 1

print(f"Total number of recorded days: {total_days}")
```

## **Task:**
### **Write Organized Data to a New CSV File**
Create a new file named organized_energy_data.csv.<br>
Write the processed and formatted energy records into this file in CSV format.<br>
The original file must remain unchanged.



```python
import csv

# Open original file for reading
with open("energy_data.csv", "r") as infile:
    reader = csv.reader(infile)
    header = next(reader)  # Skip header

    # Read all data into a list
    data = [row for row in reader]
```

```python
# Create a new CSV file and write organized data
with open("organized_energy_data.csv", "w", newline="") as outfile:
    writer = csv.writer(outfile)

    # Write header
    writer.writerow(["Serial No", "Date", "Energy Consumed (kWh)"])

    # Write processed data with serial numbers
    for idx, row in enumerate(data, start=1):
        writer.writerow([idx, row[0], row[1]])

print("organized_energy_data.csv created successfully!")
```
