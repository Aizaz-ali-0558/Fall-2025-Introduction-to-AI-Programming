# Task 1:
During charging, the battery slowly increases its percentage. Students want to simulate this charging process in code.

Keep printing "Charging..." until battery percentage reaches 100%.
-	Start with battery = 0, increase by 10 in each loop, stop when 100.


```python
battery = 0

while battery < 100:
    print("Charging...")
    battery += 10

print("Battery fully charged!")
```

# Task 2:
Engineers use countdown timers for testing. Students must simulate a timer that runs before starting a test.

Countdown timer from 10 to 0, showing "Test Running".
-	Start variable = 10, decrease each loop, stop at 0.


```python
Start_variable=10
while Start_variable>0:
    print('Test Running')
    Start_variable-=1
print('Stop')
```

# Task 3:
A voltage monitoring system asks for user input repeatedly until the system is turned off.
Keep asking user for voltage until they enter 0.
-	Use input() in while loop, stop when entered value is 0.


```python
voltage_monitoring_system=int(input('enter voltage value: '))
while voltage_monitoring_system!=0:
    print("voltage recorded",voltage_monitoring_system)
    voltage_monitoring_system=int(input('enter voltage value: '))
print('stop recording')
```

# Task 4:
Power in a system doubles when load increases. Students need to simulate this doubling until it reaches a limit.
Keep doubling the power until it exceeds 1000 W.
-	Start power = 1, multiply by 2 each loop, stop when >1000.


```python
power = 1

while power <= 1000:
    print("Power:", power, "W")
    power *= 2

print("stop power:", power, "W")
```

Start from cycle = 1 and keep printing "Cycle Running" until cycle reaches 5.
