### **Task 1:** Filter Voltage Readings
You have a list of voltage readings from a sensor:<br>
voltages = [224, 198, 231, 177, 205, 189, 240]<br>
Create a list comprehension to extract only the readings below 200 volts.<br>


```python
voltages = [224, 198, 231, 177, 205, 189, 240]
R_B_200=[x for x in voltages if x<200]
R_B_200
```

## **Task 2**: Convert Power Ratings to kW
Given power values in watts:<br>
powers_w = [50, 120, 300, 750, 1500]<br>
Use a list comprehension to convert all values to kilowatts.<br>


```python
powers_w = [50, 120, 300, 750, 1500]
Powers_in_kwh=[(x/1000) for x in powers_w]
Powers_in_kwh
```

## **Task 3:** Square Only Even Resistances
You measured resistance values:<br>
resistors = [10, 15, 22, 47, 50, 68]<br>
Create a list of squares of only the even resistor values.


```python
resistors = [10, 15, 22, 47, 50, 68]
even_resistor_square=[x**2 for x in resistors if x%2==0]
even_resistor_square
```

## **Task 5:** Extract Appliance Names Starting With ‚ÄòF‚Äô
Given appliance names:<br>
appliances = ["Fan", "Fridge", "Light", "Freezer", "Pump"]<br>
