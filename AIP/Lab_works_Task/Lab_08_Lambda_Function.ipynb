6. Lambda With reduce()
-	Use reduce + lambda to combine all numbers in a list.
-	Use reduce + lambda to join multiple strings.
-	Use reduce + lambda to find a custom smallest or largest element.


from functools import reduce

reduce(function, iterable[, initializer])
function: a function of two arguments (like lambda x, y: x + y).

iterable: the list, tuple, etc. you want to reduce.

initializer (optional): starting value for the reduction.

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

sum_numbers = reduce(lambda x, y: x + y, numbers)
print(sum_numbers)
```

```python
multiple_string=['My','Name','is','Aizaz']
sentence=reduce(lambda x,y:x+' '+y,multiple_string)
sentence
```

```python
numbers = [10, 5, 20, 8, 15]

largest = reduce(lambda x, y: x if x > y else y, numbers)

print(largest)
```

Step by step execution:

x=3, y=7 ‚Üí 3 if 3>7 else 7 ‚Üí 7

x=7, y=2 ‚Üí 7 if 7>2 else 2 ‚Üí 7

x=7, y=9 ‚Üí 7 if 7>9 else 9 ‚Üí 9

x=9, y=5 ‚Üí 9 if 9>5 else 5 ‚Üí 9

At the end, reduce() returns 9, the largest number.

8. Replace a small def with a lambda and compare readability

Given this function:

def label_meter(id):
    
    return "Meter-" + str(id).zfill(3)

Rewrite it using a lambda.

Call both versions with 12.

Write one short sentence on which is clearer.


```python
def label_meter(id):
    return "Meter-" + str(id).zfill(3)

# Test
print(label_meter(12))
```

9. Sort electrical components using lambda key

Sort this list of components by their watt rating using sorted() and a lambda key:

components = [("LED", 5), ("Fan", 60), ("Motor", 350), ("Bulb", 40)]

Then show the same sorting using operator.itemgetter.

State one advantage of itemgetter.


> sorted_by_watt_lambda = sorted(components, key=lambda x: x[1])

> print(sorted_by_watt_lambda)

```python
from operator import itemgetter
```

```python
components = [("LED", 5), ("Fan", 60), ("Motor", 350), ("Bulb", 40)]
sorted_by_watt_lambda = sorted(components, key=lambda x: x[1])

print(sorted_by_watt_lambda)
```

```python
sorted_by_watt_itemgetter = sorted(components, key=itemgetter(1))

print(sorted_by_watt_itemgetter)
```

itemgetter is faster and cleaner.

itemgetter(1)==lambda x: x[1]
