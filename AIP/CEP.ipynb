1.	Save the ‚ÄúEquipment Electric Power [kWh]‚Äù as the building‚Äôs electricity consumption in a list named **non-shiftable load.** Calculate the net sum, and it will be the building‚Äôs net consumption. 

```python
import csv
```

```python
import csv

non_shifted_load = []

with open('citylearn_challenge_2022_phase_all/Building_2.csv') as file:
    reader = csv.reader(file)
    for index, R_1 in enumerate(reader):
        if index == 0:  # skip header row
            continue
        non_shifted_load.append(float(R_1[7]))  # append 7th column
```

```python
Net_sum=sum(non_shifted_load)
print(f'The Net Consumption of The Bulding 2 is: {Net_sum} kWh')
```
